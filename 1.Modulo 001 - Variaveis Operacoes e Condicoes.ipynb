# 1.1.Variáveis

Variáveis são espaços reservados na memória que amarzenam algum dado, e é usado para realizar operações.


```python
# 1.1.1.Atrubuições de variáveis
# Antes de tudo, é importante dizer que as variáveis devem seguir algumas convenções.
"""
1.As variáveis são caseSensitive, ou seja, nomes de variáveis escritas em caixa
alta ou baixa são tratadas como diferentes.
"""
altura = 10
Altura = 5
ALTURA = 2

print(f"altura: {altura}")
print(f"Altura: {Altura}")
print(f"ALTURA: {ALTURA}")
```

## 1.3.1.Estrutura condicional if, elif

```python
# A ordem importa
"""
O codigo verifica as condições de cima para baixo. Assim que uma condição for
verdadeira, ele executa só aquela e ignora as outras.
"""

# elif significa else if
"""
É uma forma de dizer: "se a primeira for verdadeira, então verifica essa outra."
"""

# Você precisa colocar uma condição em cada if e elif
"""
Não pode deixar eles vazios. Se fizer isso o código dá erro.
"""

# Os parenteses são opcionais
"""
Você pode escrever if x > 10: ou if(x > 10): - Os dois funcionam iguais.
"""

x = 7
y = 14

if (2*x == y):
  print("y é o dobro de x")
elif (x**2 == y):
  print("y é a raiz quadrada de x")
else:
   ("y não é o dobro de x")
```

```python
# Exercício 4: Calculadora Simples

a = float(input("Digite o primeiro numero: "))
b = float(input("Digite o segundo numero: "))
operacao = input("Digite o operador: [+][-][*][/]")

if (operacao == "+"):
  print(f"{a} + {b} = {a+b}")
elif (operacao == "-"):
  print(f"{a} - {b} = {a - b}")
elif (operacao == "*"):
  print(f"{a} * {b} = {a * b}")
elif (operacao == "/"):
  print(f"{a} / {b} = {a / b}")
else:
  print("Operação inválida")
```

```python
## 1.3.2.Switch Cases
"""
Pyrhon não tem Switch Case como Java e C
Mas podemos simular usando a estrutura de dicionários
Os dicionários guardam:
  Chaves: que são as condições
  Valores: são ações

Cada chave representa um caso, e cada valor é o que deve acontecer nesse caso.

Se precisar de várias ações você pode criar funções e colocar elas como valores
no dicionário

O caso padrão (default) é feito usando o método get() do dicionário, passando
um valor padrão.
"""

def caso1():
  return "Voce escolheu a opção 1"

def caso2():
  return "Você escolheu a opção 2"

def caso3():
  return "Você escolheu a opção 3"

# Dicionario simulando Case Switch

switch = {
    1: caso1,
    2: caso2,
    3: caso3
}

opcao = int(input("Digite a opcao 1-3: "))

# Exeuta a função correspondente ou mostra padrão

resultado = switch.get(opcao, lambda: "Opcao Invalida")()
print(resultado)
```

## Exercício Switch Case

```python
# Exercício 1: Dias da semana

dias = {
    1: "Domingo",
    2: "Segunda-feira",
    3: "Terça-feira",
    4: "Quarta-feira",
    5: "Quinta-feira",
    6: "Sexta-feira",
    7: "Sábado"
}

"""
O que é um dicionário?
  O dicionário em Python é uma estrutura que associa chave e valor.
  As chaves são os números de 1 a 7.
  Os valores são os nomes dos dias da semana correspondentes.
  Por exemplo:
    ao pegar a chave 1, o valor será "Domingo"
    ao pegar a chave 3, o valor será "Terça-feira"
"""

numero = int(input("Digite um número de 1 a 7: "))

"""
input() exibe a mensagem e aguarda o usuário digitar um valor.
int() converte o valor digitado pelo usuário em um número inteiro.
O valor digitado é armazenado na variável numero.
"""

print(dias.get(numero, "Dia inválido"))

"""
O .get busca uma chave no dicionário:
  - Se a chave existe, retorna o valor correspondente.
  - Senão, retorna o valor padrão ("Dia inválido").
"""
```
